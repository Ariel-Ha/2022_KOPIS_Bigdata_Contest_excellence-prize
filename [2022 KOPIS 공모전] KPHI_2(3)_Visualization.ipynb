```python
import matplotlib.pyplot as plt

plt.rc('font', family='NanumBarunGothic')
```

```python
# 시설특성 별 wtp 평균값
wtp_mean = data_viz.groupby('시설특성')['wtp'].mean().reset_index().set_index('시설특성', drop= True)  

pal = ["#B9C0C9", "#9F8A78","#F3AB60"]
pl = sns.barplot(x= wtp_mean.index, y=wtp_mean.wtp, palette= pal)
pl.set_title("시설특성 별 wtp 평균값")
plt.show()
```

```python
# 시설특성 별 wtp 중앙값
wtp_median = data_viz.groupby('시설특성')['wtp'].median().reset_index().set_index('시설특성', drop= True)  

pal = ["#B9C0C9", "#9F8A78","#F3AB60"]
pl = sns.barplot( x= wtp_median.index, y= wtp_median.wtp, palette= pal)

pl.set_title("시설특성 별 wtp 중앙값")
plt.show()
```

```python
# 시설 특성별 내한 공연 여부 
pal = ["#B9C0C9", "#9F8A78","#F3AB60"]
pl = sns.countplot(x=data_viz["시설분류"], hue= data_viz['내한공연 여부'],palette= pal)
pl.set_title("시설특성 별 내한 공연 여부")
plt.show()
```

```python
# 시설 특성별 연령 그룹 : 2(20-40세) 파이차트 
data_age = data_viz[data_viz['나이대'] == 2]
data_age = data_age.groupby('시설분류')['나이대'].count().reset_index().set_index('시설분류', drop= True)

pal = ["#B9C0C9", "#9F8A78","#F3AB60"]
pal2 = ['yellowgreen', 'lightskyblue', 'lightcoral']
plt.figure(figsize=(7,7))
plt.pie(data_age['나이대'], labels=data_age.index, colors = pal, autopct='%.1f%%',  startangle=90, textprops = {'fontsize' : 14},counterclock=False) 
plt.show()
```

```python
# 시설 특성별 연령 3(50-70세) 파이차트   

data_age = data_viz[data_viz['나이대'] == 3]
data_age = data_age.groupby('시설분류')['나이대'].count().reset_index().set_index('시설분류', drop= True)

pal = ["#B9C0C9", "#9F8A78","#F3AB60"]
pal2 = ['yellowgreen', 'lightskyblue', 'lightcoral']
plt.figure(figsize=(7,7))
plt.pie(data_age['나이대'], labels=data_age.index, colors = pal, autopct='%.1f%%',  startangle=90, textprops = {'fontsize' : 14},counterclock=False) 
plt.show()

```

```python
# 시설 특성별 연령 4 (70-90세) 파이차트   

data_age = data_viz[data_viz['나이대'] == 4]
data_age = data_age.groupby('시설분류')['나이대'].count().reset_index().set_index('시설분류', drop= True)

pal = ["#B9C0C9", "#9F8A78","#F3AB60"]
pal2 = ['yellowgreen', 'lightskyblue', 'lightcoral']
plt.figure(figsize=(7,7))
plt.pie(data_age['나이대'], labels=data_age.index, colors = pal, autopct='%.1f%%',  startangle=90, textprops = {'fontsize' : 14},counterclock=False) 
plt.show()
```

```python
# 시설 특성별 긍부정비율 
pal = ["#B9C0C9", "#9F8A78","#F3AB60"]
pl = sns.boxplot(x = data_viz['시설분류'], y= data_viz['비율'], palette= pal)
pl.set_title("시설특성 별 기획제작사 긍부정 분포도")
plt.show()
```

```python
# 시설 특성별 오픈런 여부 
pal = ["#B9C0C9", "#9F8A78","#F3AB60"]
pl = sns.countplot(x=data_viz["시설분류"], hue= data_viz['오픈런 여부'],palette= pal)
pl.set_title("시설특성 별 오픈런 여부 분포도")
plt.show()
```
